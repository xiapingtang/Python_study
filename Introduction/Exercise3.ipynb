```python
g = (x*x for x in range(1, 6))
print (g)

for x in g:
    print (x, end='')
print('\n') 

def fib(value):
    n, a, b = 0, 0, 1
    while n < value:
        print (b, end='')
        a,b = b, a+b
        n = n+ 1
    return 'done'
fib(6)
print('\n')

def fib_generator(value):
    n, a, b = 0, 0, 1
    while n < value:
        yield b
        a,b = b, a+b
        n = n+ 1
    return 'done'
print (fib_generator(6))
for x in fib_generator(6):
    print (x, end='')

print('\n')
print (list(range(0)))

def triangles(n):
    x = [1]
    i=1
    print (x)
    while i < n:
        y = x.copy()
        y.append(1)
        for j in list(range(len(y)-2)):
            y[j+1] = x[j] + x[j+1] 
        x = y
        i = i + 1
        print (x)
        
triangles(6)
```
