在 python 用 import 或者 from...import 来导入相应的模块。

将整个模块(somemodule)导入，格式为： import somemodule

从某个模块中导入某个函数,格式为： from somemodule import somefunction

从某个模块中导入多个函数,格式为： from somemodule import firstfunc, secondfunc, thirdfunc

将某个模块中的全部函数导入，格式为： from somemodule import *

## Fibonaci series

```python
a,b = 0,1

while b <1000:
    print (b, end=' ')
    a,b = b, a+b
    
n = int(input('\n input a integer '))

a = {0:0, 1:1}
def fabonac(n):
    if n in a:
        return a[n]
    else:
        temp = fabonac(n-1) + fabonac(n-2)
        a[n] = temp
        return temp
    
for i in range(n):
    print (fabonac(i), end=' ')

del i

print ("\n 0-100 ", sum(range(101)))

a = ['Google', 'Baidu', 'Runoob', 'Taobao', 'QQ']
for i in range(len(a)):
    print(i, a[i])

for n in range(2, 10):
    for x in range(2, n):
        if n % x == 0:
            print(n, '等于', x, '*', n//x)
            break
    else:
        # 循环中没有找到元素
        print(n, ' 是质数')
        
def paixu(array) :
    for i in range(len(array)):
        for j in range(0, len(array)-i - 1):
            if array[j+1] < array[j]:
                array[j], array[j+1] = array[j+1], array[j]
        print (array)
        
    return array
                
print (paixu([41,23344,9353,5554,44,7557,6434,500,2000]))
            

import random
number = random.randint(0, 5)
guess = -1
i=0

while number != guess and i < 3:
    i+= 1;
    guess = int(input("\ninput guess number ")) 
    if guess == number:
        print ('congratulation')
    elif guess > number:
        print ('guess is too large')
    else :
        print ('guess is too small')

print ("end of game")
```
